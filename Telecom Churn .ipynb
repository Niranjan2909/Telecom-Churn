```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn import model_selection
from sklearn import metrics
from sklearn.model_selection import KFold
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier
#from sklearn.ensemble import VotingClassifier
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.ensemble import BaggingClassifier
Lr_model = LogisticRegression()
LR_bag=BaggingClassifier(base_estimator=Lr_model,n_estimators=25,random_state=0)
nb_model = GaussianNB()
NB_bag=BaggingClassifier(base_estimator=nb_model,n_estimators=24,random_state=0)
knn = KNeighborsClassifier(n_neighbors=7, weights='distance')
KNN_bag=BaggingClassifier(base_estimator=knn,n_estimators=80,random_state=0)
DT_reg=DecisionTreeClassifier(max_depth=8,criterion='entropy',random_state=0)
DT_bag=BaggingClassifier(n_estimators=10,random_state=0)
RF=RandomForestClassifier(n_estimators=150,criterion='entropy',random_state=0)
#Boosting models
LR_boost=AdaBoostClassifier(base_estimator=Lr_model,n_estimators=170)
NB_boost=AdaBoostClassifier(base_estimator=nb_model,n_estimators=400)
RF_boost=AdaBoostClassifier(base_estimator=RF,n_estimators=100)
DT_boost=AdaBoostClassifier(n_estimators=100)
GB_boost=GradientBoostingClassifier(n_estimators=100)
```

```python
models = []
models.append(('LogisticRegression', Lr_model))
models.append(('LR_bag', LR_bag))
models.append(('Naivebayes', nb_model))
models.append(('bag_nb', NB_bag))
models.append(('knn_model', knn))
models.append(('bag_knn', KNN_bag))
models.append(('DT_regularized', DT_reg))
models.append(('DT_bagged', DT_bag))
models.append(('RandomForest', RF))
models.append(('LR_boost', LR_boost))
models.append(('NB_boost', NB_boost))
models.append(('RF_boost', RF_boost))
models.append(('DT_boost', DT_boost))
models.append(('GB_boost', GB_boost))

models
```

```python
results = []
names = []
scoring = 'accuracy'
for name, model in models:
	kfold = model_selection.KFold(n_splits=5,random_state=2)
	cv_results = model_selection.cross_val_score(model, X, Y, cv=kfold, scoring=scoring)
	results.append(cv_results)
	names.append(name)
	msg = "%s: %f (%f)" % (name, np.mean(cv_results), cv_results.var())
	print(msg)
```

```python
results = []
names = []
scoring = 'recall'
for name, model in models:
    kfold = model_selection.KFold(n_splits=5)
    cv_results = model_selection.cross_val_score(model, X, Y, cv=kfold, scoring=scoring)
    results.append(cv_results)
    names.append(name)
    msg = "%s: %f (%f)" % (name, np.mean(cv_results), cv_results.var())
    print(msg)
```

```python
from sklearn.metrics import roc_curve, auc
```

```python
kfold = KFold(n_splits=5,shuffle=True,random_state=0)
for name, model in models:
    roc_auc = []
    for train, test in kfold.split(X,Y):
        Xtrain, Xtest = X[train,:],X[test,:] 
        Ytrain,Ytest = Y[train], Y[test]
        model.fit(Xtrain,Ytrain)
        Y_predict = model.predict(Xtest)
        #cm = metrics.confusion_matrix(Ytest, Y_predict)
        fpr,tpr, _ =roc_curve(Ytest,Y_predict)
        roc_auc.append(auc(fpr, tpr))
    print('AUC scores: %0.2f (+/- %0.5f) [%s]' %(np.mean(roc_auc), np.var(roc_auc,ddof = 1), name))
 
```
